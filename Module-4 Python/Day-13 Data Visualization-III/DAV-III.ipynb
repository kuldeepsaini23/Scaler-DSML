```python
import  numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Multivariate Analysis**

```python
data = pd.read_csv('game_data_final.csv')
data.head()
```

```python
top3_publishers = data["Publisher"].value_counts().index[:3]
top3_genres = data["Genre"].value_counts().index[:3]
top3_platforms = data["Platform"].value_counts().index[:3]
```

```python
top3_data = data.loc[
  (data["Publisher"].isin(top3_publishers)) &
  (data["Genre"].isin(top3_genres)) &
  (data["Platform"].isin(top3_platforms))
]
top3_data
```

```python
'''
NNC -> Numerical, Numerical , Categorical

-> numerical + numerical -> scaterplot

and three different scatter plots for each category

- Numerical + Categorical -> boxplot -> Adding one more Numerical might not make sense
'''
```

```python
sns.scatterplot(
  x = "NA_Sales",
  y = "EU_Sales",
  hue = "Publisher",
  data = top3_data
)
```

```python

#! NNC with Barplot looks like this and we cnnot understand anything
sns.barplot(
  x = "NA_Sales",
  y = "EU_Sales",
  hue = "Publisher",
  data = top3_data
)
```

## Categorial-Categorial-Numerical


```python
'''
CCN -> Categorial, Categorical, Numerical  

-> Categorial + Categorial -> Stacked/ Dodged bar plot -> Add one more Numerical 

- Numerical + Categorical -> boxplot -> Adding one more Categorial
'''
```

```python
sns.boxplot(
  x = "Publisher",
  y = "Global_Sales",
  hue = "Genre",
  data= top3_data
)
```

# **Numerical-Numerical -Numerical**

## Bubble Chart


```python
'''
NNN -> Numerical, Numerical, Numerical  

-> Numerical + Numerical -> Scalterplot -> Third Numerical as Size
'''
```

```python
sns.scatterplot(
  x = "NA_Sales",
  y = "Global_Sales",
  size = "Rank",
  sizes = (1,200),
  data = top3_data
)
plt.legend("", frameon=False)# Hide legend
plt.show()
```

```python
sns.scatterplot(
  x = "NA_Sales",
  y = "Global_Sales",
  # size = "Rank",
  # sizes = (1,200),
  hue="Rank",
  data = top3_data
)
plt.legend("", frameon=False)# Hide legend
plt.show()
#! There are So much colors which we cannot process
#* Not the best way to represent data
```

# **JointsPlots**

```python
#! Bivariate and Univaraite analysis
sns.jointplot(
  x = "NA_Sales",
  y = "EU_Sales",
  data = top3_data
)
plt.show()
```

```python

#* Bivariate and Multivariate Analysis
sns.jointplot(
  x = "NA_Sales",
  y = "EU_Sales",
  hue= "Publisher",
  data = top3_data
)
plt.show()
```

# **Co-Realtion**

```python
'''
Correlation bettween two Numerical columns -> 
-1 to 1 ->
  - -1 -> Max Negative Correlation
  - 0 -> No Correlation
  - + 1 -> Max Positive Correlation
'''
```

```python
numerical_Columns = top3_data[['Year','Rank', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
```

```python
numerical_Columns.head()
```

```python
numerical_Columns.corr()
```

```python
sns.scatterplot(
  x = "Other_Sales",
  y = "Global_Sales",
  data = top3_data
)
# See there is relation between both sales
```

### ***HeatMap*** - Corealtion in Visulaization

```python
sns.heatmap(
  numerical_Columns.corr(),
  cmap="magma",
  annot=True
)
```
